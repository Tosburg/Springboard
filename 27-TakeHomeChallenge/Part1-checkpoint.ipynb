```python
# Changing the login_time from an int64 value to datetime

df['login_time'] = pd.to_datetime(df['login_time'])
```

```python
# Below results in a error indicating duplicate keys. 

# df.set_index('login_time', inplace=True, verify_integrity=True)
```

```python
# Running the verify_integrity generated a list of duplicate rows. 
# From here I will check 2 assumptions, assuming sequential time in the data set.

# The first assumption is that 2 individuals logged in at the exact same time.
# For this to be true the login times will follow eachother.

# The second assumption is that duplicate login_time values are incorrect and should be removed.

# Value selected is from the list of duplicate rows.

df.loc[df['login_time'] == '1970-01-02 15:08:27']
```

```python
# From the above the login_time occurs at 503 and 506 indicating a change in time between the log ons.

# Here I will examine a group of log ons around the duplicated times. I am checking for sequential time and if the values
# of lines 504 and 505 follow the sequential time

df.iloc[501:511]
```

```python
# Lines 504 and 505 are in sequential time and indicate that 506 is an incorrect login_time. The login_time is either a
# duplicate or an incorrect entry that should read 15:18:27 instead of 15:08:27

# Additional note: rows 503 and 506 above represent different 15 minute groupings based on the assignment. Therefore,
# all rows should be addressed before proceeding or it will create an inaccurate account of logins per 15 minute interval.

# Below generates a simple precentage of duplicated login_time entry values

print(np.round((len(df[df.duplicated(['login_time'], keep=False)]) / len(df)) * 100,2))
```

```python
# Dropping duplicate time values as they represent less than 2% of the total.

df1 = df.drop_duplicates(subset='login_time', keep='first')
```

```python
df1 = df1.set_index('login_time')
```

```python
df1['count'] = 1
```

```python
df1_count = df1.resample('15T').sum()
```

```python
plt.figure(figsize=(16,4))
plt.plot(df1_count,color='k')
plt.title('Logins per 15 minute interval')
plt.xlabel("Date")
plt.ylabel("Number of logins")
plt.show()
```

```python
plt.figure(figsize=(16,4))
plt.plot(df1_count.loc['1970-02-15':'1970-03-15'],color='k')
plt.title('Close Up of logins from 02-15 to 03-15')
plt.xlabel("Date")
plt.ylabel("Number of logins")
plt.show()
```
